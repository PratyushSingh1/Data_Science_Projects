```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_data.csv')
```

```python
df['school'].value_counts().plot.bar()
plt.title('School population')
plt.show()
```

```python
df['sex'].value_counts().plot.bar()
plt.title('Female vs Male(Population)')
plt.show()
```

```python
df['age'].value_counts().plot.bar()
plt.title('Age area')
plt.show()
```

```python
df['address'].value_counts().plot.bar()
plt.title('Urban VS Rular')
plt.show()
```

```python
plt.figure(figsize=(20,20))
correlation = df.corr()
sns.heatmap(correlation,xticklabels = correlation.columns, yticklabels = correlation.columns, annot=True)
```

```python
sns.histplot(x='school', hue='age',data=df)
plt.show()
```

```python
plt.figure(figsize=(14,10))
plt.subplot(2,3,1)
sns.barplot(x= 'absences', y= 'sex', errwidth=0, data=df)
plt.subplot(2,3,2)
sns.barplot(x='failures', y='Medu', errwidth=0, data=df)
plt.subplot(2,3,3)
sns.barplot(x='failures', y='Fedu', errwidth=0, data=df)
plt.subplot(2,3,4)
sns.violinplot(x='failures', y='age', data=df)
plt.subplot(2,3,5)
sns.violinplot(x='failures', y='Medu', data=df)
plt.subplot(2,3,6)
sns.violinplot(x='failures', y='Fedu', data=df)
plt.show()
```
